```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("stock.xlsx")
df=pd.DataFrame(data)
print(df)
grouped_by=df.groupby("Ticker")["Close"].describe()
print(grouped_by)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("stock.xlsx")
df=pd.DataFrame(data)
print(df)
colors=["red","blue","green","violet"]
plt.bar(df["Ticker"],df["Volume"],color=colors)
plt.xlabel("Ticker",fontsize=20)
plt.ylabel("Volume",fontsize=20)
plt.title("Stock Market",fontsize=20)
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("stock.xlsx")

df=pd.DataFrame(data)
print(df)
plt.plot(df["Date"],df["Close"],label="Ticker")
plt.xlabel("Date",fontsize=20)
plt.ylabel("Close",fontsize=20)
plt.title("Stock Market",fontsize=20)
plt.legend()
plt.show
```
